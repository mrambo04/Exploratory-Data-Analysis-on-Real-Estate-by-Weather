```python
def extract_city(location):
    if pd.isna(location):
        return None
    parts = [p.strip() for p in location.split(",")]
    return parts[-1] if parts else None

estate['City'] = estate['Location'].apply(extract_city)
```

```python
estate['City_clean'] =estate['City'].str.strip().str.lower()
weather['District_clean'] = weather['District'].str.strip().str.lower()
```

```python
realestate_weather = pd.merge(estate,weather,left_on='City_clean',right_on='District_clean', how='left')
```

```python
realestate_weather
```

```python
states_list = weather['State/UT'].dropna().unique().tolist()
states_list_lower = [s.lower() for s in states_list]
```

```python
def extract_state_from_title(title):
    title_lower = title.lower()
    for state in states_list_lower:
        if state in title_lower:
            return state.title()
    return None

disaster['State_Match'] = disaster['Title'].apply(extract_state_from_title)
```

```python
realestate_weather['State_clean'] = realestate_weather['State/UT'].str.strip().str.title()
```

```python
merged_df = pd.merge(realestate_weather,disaster,left_on='State_clean',right_on='State_Match',how='left')
```

```python
def assign_zone(location):
    location = str(location).lower()
    if "chennai" in location:
        return "coastal"
    elif "delhi" in location or "bangalore" in location or "hyderabad" in location:
        return "inland"
    elif "dehradun" in location or "shimla" in location or "manali" in location:
        return "mountain"
    else:
        return "inland"
```

```python
estate["Zone"] = estate["Location"].apply(assign_zone)
```
